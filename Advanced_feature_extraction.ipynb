### Reading ratings data

```python
movies_features
```

```python
#Reading the ratings data
ratings_filename = "/ml-1m/ratings.dat"
dfratings = pd.read_csv(my_path[0]+ratings_filename, index_col=None, sep='::', header=None)
dfratings.columns = ["userId", "movieId", "rating", "timestamp"]
```

```python
all_users = user_features["userId"]
avg_score_diff = []
for userId in all_users:
    user_ratings = dfratings[dfratings["userId"] == userId]
    num_ratings = user_ratings.shape[0]
    score_diff = 0
    for index, rating in user_ratings.iterrows():
        movieId = rating["movieId"]
        score = rating["rating"]
        movie_features = movies_features.loc[movies_features["movieId"] == movieId]
        avg_score = (movie_features["avg_rate"]).item()
        score_diff += (score - avg_score)
    avg_diff = score_diff/float(num_ratings)
    avg_score_diff.append(avg_diff)
    if (userId % 1000 == 0):
        print userId
```
